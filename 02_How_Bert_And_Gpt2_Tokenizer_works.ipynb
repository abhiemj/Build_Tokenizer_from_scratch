## [1] BERT

```python
from transformers import BertTokenizer, BertModel
tokens_from_bert = BertTokenizer.from_pretrained('bert-base-uncased')
word = "gullable"
encoded_word = tokens_from_bert(word)
print(encoded_word)
```

```python
## Get the output from these tokens , what do they represent ?
letter_0 = tokens_from_bert.convert_ids_to_tokens([101])
print(letter_0)
letter_1 = tokens_from_bert.convert_ids_to_tokens([19739])
print(letter_1)
letter_2 = tokens_from_bert.convert_ids_to_tokens([4571])
print(letter_2)
letter_3 = tokens_from_bert.convert_ids_to_tokens([3468])
print(letter_3)
letter_4 = tokens_from_bert.convert_ids_to_tokens([102])
print(letter_4)
```

## GPT 2

```python
from transformers import GPT2Tokenizer, GPT2Model
tokens_from_gpt = GPT2Tokenizer.from_pretrained('gpt2-medium')
word = 'gullable'
encoded_word = tokens_from_gpt(word)
print(encoded_word)
```

```python
## convert ids to token
words = tokens_from_gpt.convert_ids_to_tokens(ids=[70, 724, 540])
print(words)
```

## Gpt 2 is decoder only Transformer
GPT-2 as a Decoder: GPT-2, developed by OpenAI, is designed as a generative model. It operates like the decoder part of the transformer architecture. It takes an input sequence and generates the next part of the sequence based on that input. However, it doesn't have a separate encoder component; it is purely generative, predicting the next token in a sequence given the previous tokens.

Autoregressive Nature: GPT-2 is autoregressive, meaning it generates text by predicting one word at a time in a sequence. Each new word it generates is based on the sequence of words that came before it.
